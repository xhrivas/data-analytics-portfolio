# A Plant-Based Coffee Shop

Coffee is a global phenomenon. Across the world, thousands of new coffee shops open each year, with specialty shops becoming increasingly popular.

In this project I will clean a survey dataset of approximately 1000 coffee enthusiasts to identify the most popular dairy choices. 

```python
import pandas as pd
survey = pd.read_csv('coffee-survey-results.csv')
survey #display dataframe
```

Those are some wordy column names! This dataset has 30 columns that appear to be mostly full sentence questions.

### Narrowing down the columns
Let's look through all the columns to identify which ones are needed to answer the question about dairy preferences.

```python
survey.columns
```

Columns numbered 4-11 ask about dairy preferences. 

```python
needed_columns = [
    "What kind of dairy? (Whole milk)",
    "What kind of dairy? (Skim milk)",
    "What kind of dairy? (Half and half)",
    "What kind of dairy? (Coffee creamer)",
    "What kind of dairy? (Flavored creamer)",
    "What kind of dairy? (Oat milk)",
    "What kind of dairy? (Almond milk)",
    "What kind of dairy? (Soy milk)",
]
dairy = survey[needed_columns]
dairy 
```

This narrows the data to just the 8 columns about dairy preferences.

### Renaming the columns

We can shorten these column names using the `rename()` method. This method takes in map where the keys are the column names to update and the values are the new names.

```python
name_map = {
    'What kind of dairy? (Whole milk)': 'Whole milk',
    'What kind of dairy? (Skim milk)': 'Skim milk',
    'What kind of dairy? (Half and half)': 'Half and half',
    'What kind of dairy? (Coffee creamer)': 'Coffee creamer',
    'What kind of dairy? (Flavored creamer)': 'Flavored creamer',
    'What kind of dairy? (Oat milk)': 'Oat milk',
    'What kind of dairy? (Almond milk)': 'Almond milk',
    'What kind of dairy? (Soy milk)': 'Soy milk',
}
```

In the name map above, the new name for 'What kind of dairy? (Whole milk)' is 'Whole milk'. Let's use this name map to rename the column names and improve readability.


```python
dairy = dairy.rename(columns=name_map)
dairy #display the dataframe
```

These column names are much easier to work with.

### Making sense of the data

As I am cleaning a dataframe, it's important to understand what the values in each row and column mean.

Here each row represents one coffee enthusiast and their yes/no answers to each survey question. If the enthusiast adds a particular type of dairy to their coffee, that column will show a value of 1, while a value of 0 indicates that they do not use that type of dairy.

Notice that there are also several `NaN` values (Not a Number), which represents missing values. In this case, NaNs indicate that the survey question was not answered.

Let's use the `isna()` method to take a closer look at how many `NaN` values each column has.


```python
dairy.isna().sum()
```

Notice that each dairy option has the same number of NaN values. Why might this be the case?

This particular survey had a question that asks if you add any dairy to your coffee. If the respondent answered "No", the dairy part of the survey was skipped entirely.

Since these rows have no data, we can drop them using `dropna()`.


```python
dairy = dairy.dropna()
dairy #display dataframe
```

After dropping all the rows with NaNs, there are now 523 rows left in the dataframe. Many rows were removed, such as the row with index 1.

Now that we have cleaned the data, we can investigate dairy preferences.

### What dairy options are most preferred?

What percentage of people consume each type of dairy? We can use the `mean()` method on the dataframe to calculate the ratio of "yes" answers for each column. Multiplying by 100 will give us these ratios as percentages.

```python
dairy_preferences = dairy.mean() * 100
dairy_preferences #display results
```

It looks like whole milk is the most popular. A quick scan of these numbers tells us that the percentages add up to more than 100%. This happens because people often like more than one type of dairy in their coffee.

### Visualizing dairy preferences
Let's visualize the dairy preferences as a horizontal bar chart. We need to first sort `dairy_preferences` and then will use the `barh()` method from matplotlib.

```python
dairy_preferences = dairy_preferences.sort_values()
dairy_preferences #display series
```

And now we can create our horizontal bar plot.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 3))
plt.barh(dairy_preferences.index, dairy_preferences)
plt.xlabel("Percent")
```

The most popular plant based dairy choice is oat milk. This might be a great choice for the default dairy option in a specialty coffee shop.
